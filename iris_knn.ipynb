```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
#import the dataset
from sklearn import datasets
df = pd.read_csv('Iris.csv')
print(df.head(10))
x = df.iloc[:, [0,1,2,3]].values
kmeans5 = KMeans(n_clusters=3)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
plt.scatter(x[:,0],x[:,1],c=y_kmeans5,cmap='rainbow')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
#import the dataset
from sklearn import datasets
data=pd.read_csv('Iris.csv')
print(data)

#extract all rows and list of 4 columns
x=data.iloc[:,[0,1,2,3,4]].values
# K-Means clustering
usingkmeans=KMeans(n_clusters=3)
y=usingkmeans.fit_predict(x)
#fit_predict used in unsupervised machine learning
print(y)

usingkmeans.cluster_centers_
#code book
#x[start_index : end_index : step] 
plt.scatter(x[:,0],x[:,1],c=y,cmap='spring_r',alpha=0.4)
#x[:,0] x coordinates and x[:,1]y coordinates
#c can be a single color format string, or a sequence of color specifications of length N
```
